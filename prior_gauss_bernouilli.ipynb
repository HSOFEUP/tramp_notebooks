```python
%matplotlib inline
import numpy as np, pandas as pd, matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
import logging
logger = logging.getLogger()
```

## Model

- $p(x) = \rho \mathcal{N}(x) + (1-\rho)\delta_0(x)$
- infer $x$ from noisy observation $r,v$ or $a=\frac{1}{v},b=\frac{r}{v}$
- $Z = \int dx\, p(x) e^{-\frac{1}{2} a x^2 + b x}$

```python
from tramp.priors import GaussBernouilliPrior
```

## Posterior

```python
def run_posterior(rho, r, v):
    prior = GaussBernouilliPrior(size=1, rho=rho)
    a, b = 1/v, r/v
    rx, vx = prior.compute_forward_posterior(a, b)
    return dict(a=a, b=b, rx=rx, vx=vx)
```

```python
from ml_physics_utils import Task, DataFrameResultHandler
from ml_physics_utils.notebook import progressbar
from ml_physics_utils.plot import qplot
logger.setLevel(logging.ERROR)


result_handler = DataFrameResultHandler()
task = Task(run_posterior, rho=0.5, v = [0.5, 2], r = np.linspace(-10,10,101))
task.run(result_handler, on_progress=None)
df = result_handler.get_dataframe()
```

```python
qplot(df, x="r", y="rx", color="v")
```

```python
qplot(df, x="r", y="vx", color="v")
```

```python
result_handler = DataFrameResultHandler()
task = Task(run_posterior, rho=[0.25, 0.5, 0.75], v = 1, r = np.linspace(-10,10,101))
task.run(result_handler, on_progress=None)
df = result_handler.get_dataframe()
```

## Error function

```python
def run_error(rho, v):
    prior = GaussBernouilliPrior(size=1, rho=rho)
    a = 1/v
    error = prior.compute_forward_error(a)
    return dict(a=a, error=error)
```

```python
result_handler = DataFrameResultHandler()
task = Task(run_error, rho=[0.25, 0.5, 0.75], v = 10**np.linspace(-3,3,101))
task.run(result_handler, on_progress=None)
df = result_handler.get_dataframe()
```

```python
qplot(df, x="v", y="error", color="rho", xlog=True, ylim=(0,1))
```
